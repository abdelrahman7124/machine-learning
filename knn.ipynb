```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
```

```python
#load the dataset
iris = datasets.load_iris()

# Choose the features for x = sepal length & pedal length and y = species
x = iris.data[:, [0,2]] 
y = iris.target  

#splittig the dataset
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# Train the model using K-Nearest Neighbors
model = KNeighborsClassifier(n_neighbors=1)
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
```

```python
#plotting the desicion boundary using mesh grid
x_min = x[:, 0].min() - 1
x_max = x[:, 0].max() + 1
y_min = x[:, 1].min() - 1
y_max = x[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),
                     np.arange(y_min, y_max, 0.01))
Z = model.predict(np.c_[xx.flatten(), yy.flatten()])
Z = Z.reshape(xx.shape)
plt.contourf(xx, yy, Z, alpha=0.4)
plt.scatter(x[:, 0], x[:, 1],c=y)
plt.title('K-Nearest Neighbors')
plt.xlabel('sepal length')
plt.ylabel('pedal length')
plt.show()
```

```python
#calculating the accuracy
accuracy = model.score(X_test, y_test) * 100
print(f"Accuracy: {accuracy:.2f}%")
```
